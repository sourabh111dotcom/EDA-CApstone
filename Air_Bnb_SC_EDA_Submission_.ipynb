### Chart-16
###Price Group Analysis of Neighborhood Groups

By using quantile function, I divide price interval into five. Then, we define values in this intervals as "Very low", "Low", "Medium", "High" ,and "Very high". Then by using these categorical values, I prepare pie chart and for each neighborhood group.

````python
#calculating quantile values
quant = list(airbnb.price.quantile([0.2,0.4,0.6,0.8,1]))
quant
````

````python
#funtion to categorizing price in 5 groups
def group_price(x):
    try:
        if 0 <= x <= quant[0]:
            return 'Very Low'
        elif quant[0] < x <= quant[1]:
            return 'Low'
        elif quant[1]< x <= quant[2]:
            return 'Medium'
        elif quant[2] < x <= quant[3]:
            return 'High'
        else:
            return 'Very High'
    except:
        return np.nan
````

````python
#applying on price column
airbnb['price_group'] = airbnb['price'].apply(group_price)
````

````python
airbnb['price_group']
````
